```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample data
data = {
    "Day": [1, 2, 3, 4, 5],
    "Temperature": [30, 32, 35, 33, 31]
}

df = pd.DataFrame(data)

# Show data
print(df)

# Plot graph
plt.plot(df["Day"], df["Temperature"])
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Temperature Analysis")
plt.show()
```
